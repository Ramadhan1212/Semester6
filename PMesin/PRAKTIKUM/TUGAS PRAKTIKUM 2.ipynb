```python
path = "covid19.csv"
df = pd.read_csv(path)
```

```python
df.head()
```

```python
kasus1 = df[df['Country/Region'] == 'Indonesia']
kasus1
```

```python
kasus2 = df[df['Country/Region'] == 'Japan']
kasus2
```

```python
kasus_ind = kasus1[kasus1['Confirmed'] > 0]
kasus_ind
```

```python
kasus_jpn = kasus2[kasus2['Confirmed'] > 0]
kasus_jpn
```

```python
#Menambahkan date base untuk indonesia dan japan

kasus_ind['Date'] = pd.to_datetime(kasus_ind['Date'])
kasus_jpn['Date'] = pd.to_datetime(kasus_jpn['Date'])
kasus_ind
```

```python
kasus_ind['BaseDate'] = pd.to_datetime('2020-03-02')
kasus_ind.head()
```

```python
kasus_jpn['BaseDate'] = pd.to_datetime('2020-01-22')
kasus_jpn.head()
```

```python
kasus_jpn
```

```python
#selisih tanggal 

kasus_ind['HariKe'] = (kasus_ind['Date'] - kasus_ind['BaseDate'])
kasus_jpn['HariKe'] = (kasus_jpn['Date'] - kasus_jpn['BaseDate'])
kasus_ind
```

```python
kasus_ind.HariKe = kasus_ind.HariKe/pd.Timedelta(1,  unit='d')
kasus_ind.head()
```

```python
kasus_jpn
```

```python
kasus_jpn.HariKe = kasus_jpn.HariKe/pd.Timedelta(1,  unit='d')
kasus_jpn.head()
```

```python
plt.plot(kasus_ind.HariKe, kasus_ind.Confirmed)
plt.plot(kasus_jpn.HariKe, kasus_jpn.Confirmed)
plt.xlabel('Hari ke-')
plt.ylabel('Jumlah kasus')
plt.title('SELISIH ANGKA COVID JEPANG DAN INDONESIA')
plt.legend(['Indonesia', 'Jepang'])
plt.show
```
